```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinical_df = pd.read_csv(clinical_trial_data_to_load)
# Combine data into a single dataset
mouse_clinical_df = mouse_df.merge(clinical_df, how='outer', on='Mouse ID')
# Rename columns
mouse_clinical_df.columns = ['MouseID','Drug','Timepoint','Tumor Volume (mm3)','Metastatic Sites']
# Display the data table for preview
mouse_clinical_df.sort_values('MouseID').head(5)
# Errorbar plot generator
def creat_errorbar_plot(df_pivot, plot_name, plot_title, plot_xlabel, plot_ylabel, plot_ylim):
    plot_df = df_pivot.reset_index()
    dics = []
    plot_x = plot_df['Timepoint']
    plot_columns = ['Capomulin','Infubinol','Ketapril','Placebo']
    plot_colors = ['red', 'blue', 'green', 'black']
    plot_markers= ['o',6,'s','d']
    plot_error = [plot_df[plot_columns[i]].sem() for i in range(4)]
    
    fig, plot_name = plt.subplots(figsize=(8,5))
    fig.suptitle(plot_title, fontsize='14')
    for i,drug in enumerate(plot_columns):
        #Tumor Volume mm3 is Y axis
        plot_name.scatter(plot_x, plot_df[drug], color=plot_colors[i], 
                          marker='o', edgecolors='r')
        
        #Tumor Volume mm3 is X axis
        #plot_name.scatter(plot_df[drug], plot_x, color=plot_colors[i], 
        #                  marker=plot_markers[i], edgecolors='r')

    plot_name.grid()
    plot_name.set_xlabel(plot_xlabel)
    plot_name.set_ylabel(plot_ylabel)
    plot_name.set_ylim(plot_ylim)
    plot_name.legend(loc='best')
    
    return plt.show()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
trt_mean_df = mouse_clinical_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean().reset_index()
# Convert to DataFrame

# Preview DataFrame
trt_mean_df.head(5)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
trt_std_error_df = mouse_clinical_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem().reset_index()
# Preview DataFrame
trt_std_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
trt_minor_df = mouse_clinical_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean().reset_index()
trt_minor_df_pivot = trt_minor_df.pivot(
    columns='Drug',
    index = 'Timepoint',
    values = 'Tumor Volume (mm3)'
)
# Preview that Reformatting worked
trt_minor_df_pivot.head()
```

```python
#Generate the Plot (with Error Bars)
creat_errorbar_plot(trt_minor_df_pivot, 'trt_plot', 'Tumor Response to Treatment', 'Time (Days)', 'Tumor Volume (mm3)', [33,76])
#creat_errorbar_plot(trt_minor_df_pivot, 'trt_plot', 'Tumor Response to Treatment', 'Time (Days)', 'Tumor Volume (mm3)', [76,33])
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_dr = mouse_clinical_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean().reset_index().set_index(['Drug','Timepoint'])
# Preview DataFrame
met_dr.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_std_dr = mouse_clinical_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem().reset_index().set_index(['Drug','Timepoint'])
# Preview DataFrame
met_std_dr.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_minor_df = mouse_clinical_df.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean().reset_index()
# Create Pivot
met_minor_df_pivot = met_minor_df.pivot(
columns='Drug',
    index='Timepoint',
    values='Metastatic Sites'
)
# Preview that Reformatting worked
met_minor_df_pivot.head(5)
```

```python
#Generate the Plot (with Error Bars)
creat_errorbar_plot(met_minor_df_pivot, 'met_plot', 'Metastatic Spread During Treatment', 'Treatment Duration (Days)', 'Met. Sites',  [-0.2,3.8])
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rates_df = mouse_clinical_df.groupby(['Drug','Timepoint']).MouseID.count().reset_index()
survival_rates_df.columns = ['Drug', 'Timepoint', 'Mouse Count']
# Preview DataFrame
survival_rates_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rates_df_pivot = survival_rates_df.pivot(
columns='Drug',
    index='Timepoint',
    values='Mouse Count'
)
# Preview the Data Frame
survival_rates_df_pivot.head(15)
```

```python
# Generate the Plot (Accounting for percentages)
creat_errorbar_plot((survival_rates_df_pivot / survival_rates_df_pivot.iloc[0])*100, 'survival_rates_plot', 'Survival During Treatment', 'Time (Days)', 'Survival rate(%)',  [33,104])
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
summary = trt_minor_df_pivot.iloc[9] / trt_minor_df_pivot.iloc[0] - 1
print(summary)
# Display the data to confirm
summary = pd.DataFrame(summary).reset_index()
summary.columns = ['Drug','Percentage_change']
relevant_percentage = (summary[(summary.Percentage_change < 0)])
```

```python
# Store all Relevant Percent Changes into a Tuple
summary_plot_df = summary.loc[[0,2,3,5]]
summary_plot_df['color'] = summary_plot_df.Percentage_change.apply(lambda x: 'g' if x < 0 else 'r')
summary_plot_df = summary_plot_df.reset_index()
```

```python
fig, summary_plot = plt.subplots(figsize=(8,5))
fig.suptitle('Tumor Response to Treatment', fontsize='14')
summary_plot.bar(summary_plot_df.Drug, summary_plot_df.Percentage_change *100, color =summary_plot_df['color'], width=1)
for i in range(len(summary_plot_df)):
    if i > 0:
        y_modify = 5
    else:
        y_modify = -5
    plt.text(x = 0 + i, 
             y = y_modify, 
             s = f"{int(summary_plot_df.Percentage_change.iloc[i]*100)}%",
             size = 12,
             color = 'white',
             ha = 'center',)
plt.grid()
summary_plot.set_ylabel('% Tumor Volume Change')
plt.show()
```
